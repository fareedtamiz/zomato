# Zomato Data Analysis 
 The project titled Zomato Data Analysis involves analyzing a dataset of restaurant information from Zomato. Here's a summary of its key components:

1. Libraries Imported: The project makes use of libraries such as pandas, numpy, matplotlib, and seaborn for data handling and visualization. 
2. Dataset Loading: The dataset is read from a CSV file called "Zomato data.csv", which includes columns such as restaurant name, online order availability, table booking options, ratings, votes, cost for two, and cuisine type. 
3. Data Preparation: 
The dataset is previewed using the head() function to understand the initial structure. 
The "rate" column, which contains ratings in the format "4.1/5", is converted to a float by removing the denominator. 
4. Objective: The project appears to be focused on cleaning, processing, and analyzing the Zomato dataset to extract insights, such as restaurant ratings, online ordering trends, and cost distribution.

```python
# Importing important libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('Zomato data .csv')
```

```python
data.head()
```

```python
#Let's first convert the "rate" column to a float data type and remove the denominator.
def handleRate(value):
    value=str(value).split('/')
    value=value[0];
    return float(value)
 
data['rate']=data['rate'].apply(handleRate)
print(data.head())
```

```python
# Now checking the missing values in data
data.info()
```

 __Above info state that there is no missing values in it.__

```python
#Let’s explore the listed_in (type) column
sns.countplot(x=data['listed_in(type)'])
plt.xlabel("Type of restaurant")
```

__The above plot shows that majority of customer preference is Dining and after that Cafes.__

```python
grouped_data = data.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': grouped_data})
plt.plot(result, c="maroon", marker="o")
plt.xlabel("Type of restaurant", c="blue", size=20)
plt.ylabel("Votes", c="yellow", size=20)
```

__A greater number of people prefer dining restaurants.__








```python
#Let's take a closer look at the online_order column.
```

__This indicates that most restaurants do not accept online orders.__

```python
#Let’s explore the rate column.
plt.hist(data['rate'],bins=5,color='brown')
plt.title("Ratings Distribution")
plt.show()

```

__The majority of restaurants received ratings ranging from 3.5 to 4.__

```python
#Let’s explore the approx_cost(for two people) column.

couple_data=data['approx_cost(for two people)']
sns.countplot(x=couple_data)
```

__The majority of couples prefer restaurants with an approximate cost of 300 rupees.__

```python
#Now we will examine whether online orders receive higher ratings than offline orders.

plt.figure(figsize = (6,6))
sns.boxplot(x = 'online_order', y = 'rate', data = data)
```

__Offline orders received lower ratings in comparison to online orders, which obtained excellent ratings.__

__Dining restaurants primarily accept offline orders, whereas cafes primarily receive online orders. This suggests that clients prefer to place orders in person at restaurants, but prefer online ordering at cafes.__

```python

# Assuming the data has columns like 'listed_in(type)' and 'online_order'
# Replace 'data' with the actual DataFrame name
pivot_table = data.pivot_table(index='listed_in(type)', columns='online_order', aggfunc='size', fill_value=0)

# Plot the heatmap
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", fmt='d')
plt.title("Online Order Acceptance by Dining Type")
plt.xlabel("Online Order (Yes/No)")
plt.ylabel("Dining Type")
plt.show()
```

```python
# Grouping data by dining type and counting the number of orders for online and offline
order_counts = data.groupby(['listed_in(type)', 'online_order']).size().unstack()

# Plotting a stacked bar chart
order_counts.plot(kind='bar', stacked=True, color=['skyblue', 'orange'], figsize=(10, 6))

# Adding labels and title
plt.title("Online vs Offline Orders by Dining Type")
plt.xlabel("Dining Type")
plt.ylabel("Number of Orders")
plt.legend(["Offline Orders", "Online Orders"], title="Order Type")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Show the plot
plt.show()
```

__A stacked bar chart where each bar represents a dining type, and the sections of the bar represent the counts of online vs. offline orders for that particular type of establishment.__

```python
max_votes = data['votes'].max()
restaurant_with_max_votes = data.loc[data['votes'] == max_votes, 'name']

print("Restaurant(s) with the maximum votes:")
print(restaurant_with_max_votes)
```

__Above Output explains the restaurant name with maximum voting.__

```python
# Filter data to only include rows where online_order is 'Yes'
online_order_data = data[data['online_order'] == 'Yes']

# Group by dining type and count the number of establishments offering online orders
order_counts = online_order_data['listed_in(type)'].value_counts()

# Plotting a bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=order_counts.values, y=order_counts.index, palette='viridis')

# Adding labels and title
plt.title("Most Popular Restaurant Types for Online Orders")
plt.xlabel("Number of Establishments")
plt.ylabel("Restaurant Type")

# Show the plot
plt.show()
```
